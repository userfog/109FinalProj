```python
NUM_PLAYERS_PATH = "../../num_players_to_df/"

def nice_graph(xlab, ylab, title):
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
```

# 3 Betting Pre Flop

```python
# Helper function for defining whether or not a player paised
def conatians_raise_or_all_in(s):
    return "r" in s or "A" in s

# Filter by three bet Pre Flop
def three_bet_pre_flop(df):
    df = df.sort(["pos"])
    two_raises = 2
    for ind, row in df.iterrows():
        if conatians_raise_or_all_in(row.prflop):
            two_raises = two_raises - 1
        if two_raises == 0:
            return row
    return None


# Create Dictionary From Number of Players to DataFrame
def make_num_players_to_df(r=range(2,11)):
    d = dict.fromkeys(r)
    for i in r:
        d[i]  = pd.DataFrame.from_csv((NUM_PLAYERS_PATH + "num_players_%d.csv" % i))
        d[i].reset_index()
    return d

d = make_num_players_to_df()
d[2].head()
```

## Helper Functions for Calculating Average Winning Per Position

```python
"""
Function which plots multiple bar graphs
Given a list of tuples (x,y) make bar graphs
"""
def matrix_bar_plot(list_of_tuples, aspect):
    fg = plt.figure(figsize=(8,10))
    adj = plt.subplots_adjust(hspace=1,wspace=1)
    sp = plt.subplot(aspect[0],aspect[1],1)
    for idx, (x, y) in enumerate(list_of_tuples):
        ln = plt.subplot(aspect[0],aspect[1], idx+1)
        plt.bar(x,y)
        nice_graph("Position", "Expected Value", ("EV vs Pos (%d)" % max(x)))
        pylab.xticks(x, map(str, x))

"""
Given a DataFrame, retuns a list of tuples (player_postion, averag_winnings)
"""
def get_pos_vs_avgwin(d):
    graphs = [None]*len(d)
    i = 0
    for key, value in d.iteritems():
        value["real_winnings"] = value.winnings - value.action
        positions = range(int(min(value.pos.tolist())),int(max(value.pos.tolist()))+1)
        p = value.groupby("pos")
        avg_winnings = p.mean().real_winnings.tolist()
        graphs[i] = (positions, avg_winnings)
        i+=1
    return graphs
```

## Average Winning Per Position All Players

```python
graphs = get_pos_vs_avgwin(d)
graphs = sorted(graphs, key=lambda x:max(x[0]))
matrix_bar_plot(graphs, [3,3])
plt.show()
```

## Average winning per position 3 betters

```python
three_betters = dict.fromkeys(d.keys())

graphs = [None]*len(d)
i = 0
for key, value in d.iteritems():
    tmp = value.groupby("timestamp").apply(three_bet_pre_flop)
    tmp = tmp.dropna(thresh=1)
    three_betters[key] = tmp

graphs = get_pos_vs_avgwin(three_betters)
graphs = sorted(graphs, key=lambda x:max(x[0]))
matrix_bar_plot(graphs, [3,3])
plt.show()
```
