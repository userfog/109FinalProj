3 Betting Pre Flop

```python
def conatians_raise_or_all_in(s):
    return "r" in s or "A" in s

def three_bet_pre_flop(df):
    df = df.sort(["pos"])
    two_raises = 2
    for ind, row in df.iterrows():
        if conatians_raise_or_all_in(row.prflop):
            two_raises = two_raises - 1
        if two_raises == 0:
            return row
    return None
    
test_rounds = pd.read_csv('../../num_players_to_df/all_players.csv')
test_rounds = test_rounds.drop(["Unnamed: 0","Unnamed: 0.1"], axis=1)
test_rounds = test_rounds.groupby("timestamp").apply(three_bet_pre_flop)
test_rounds = test_rounds.dropna(thresh=1)
test_rounds.head(n=50)
```

Average Winning Per Position

```python
test_rounds["real_winnings"] = test_rounds.winnings - test_rounds.action
p = test_rounds.groupby("pos")
plt.bar(xrange(2,11), p.mean().real_winnings)
```
